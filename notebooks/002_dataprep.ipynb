# Encontrando e removendo outliers

```python
#Tukey's method
def tukeys_method(df, variable):
    #Takes two parameters: dataframe & variable of interest as string
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3-q1
    inner_fence = 1.5*iqr
    
    #inner fence lower and upper end
    inner_fence_le = q1-inner_fence
    inner_fence_ue = q3+inner_fence
    
    return df[(df[variable] > inner_fence_le) & (df[variable] < inner_fence_ue)]
```

```python
df_filter = df.copy()

columns = ['Age', 'Income']

for var in columns:
    df_filter = tukeys_method(df_filter, var)
    
df_filter
```

```python
df_filter.to_csv('../data/filtered_data.csv', index=False)
```
